### 모의고사 1회 문제 - 2유형

```python
# 1. 필요한 패키지 임포트
import numpy as np
import pandas as pd
import sklearn

# 2. 데이터 불러오기
dat = pd.read_csv('https://raw.githubusercontent.com/YoungjinBD/dataset/main/Bank_Personal_Loan_Modeling.csv')
dat_processed = dat.dropna().drop(['ID', 'ZIP Code'], axis = 1, inplace = False)

X = dat_processed.drop(['Personal Loan'], axis = 1)
y = dat_processed['Personal Loan']
```

```python
# 3. 데이터 탐색
X.info()
X.describe()
```

```python
# 4. 데이터 전처리
from sklearn.preprocessing import Normalizer

scaler = Normalizer()
X_transform = scaler.fit_transform(X)
```

```python
# 5. 데이터 분석 준비 
from sklearn.model_selection import train_test_split
# 계층을 유지하라고 했으므로 stratify = y / trainsize = 0.7
X_train, X_test, y_train, y_test = train_test_split(X_transform, y, stratify = y, train_size = 0.7, random_state = 1234)

print(X_train.shape)
```

```python
# 6. 데이터 분석 수행

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy = 0
nn_best = 0
for nn in range(1, 25):
    knn = KNeighborsClassifier(n_neighbors = nn)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    acc = accuracy_score(y_test, pred)
    accuracy = max(accuracy, acc)

    if accuracy == acc:
        nn_best = nn

print(nn_best)
print(accuracy) # 0.926 으로 k = 5 일때 가장 좋다. 
```
